```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
# Load distilbert results
df_distilbert = pd.read_csv("../data/processed/distilbert_IMDB_movie_reviews.csv", encoding="cp1252")
true_labels_distilbert = df_distilbert["sentiment"]
predicted_labels_distilbert = df_distilbert["predicted sentiment"]
```

```python
# Create confusion matrix and classification report for DistilBERT
cm_distilbert = confusion_matrix(true_labels_distilbert, predicted_labels_distilbert)
cr_distilbert = classification_report(true_labels_distilbert, predicted_labels_distilbert)
```

```python
# Load roberta results
df_roberta = pd.read_csv("../data/processed/roberta_IMDB_movie_reviews.csv", encoding="cp1252")
true_labels_roberta = df_roberta["sentiment"]
predicted_labels_roberta = df_roberta["predicted sentiment"]
```

```python
# Create confusion matrix and classification report for RoBERTa
cm_roberta = confusion_matrix(true_labels_roberta, predicted_labels_roberta)
cr_roberta = classification_report(true_labels_roberta, predicted_labels_roberta)
```

```python
print("DistilBERT classification report:\n")
print(cr_distilbert)

print("RoBERTa classification report:\n")
print(cr_roberta)
```

```python
# Plot confusion matrices next to each other
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# DistilBERT
disp1 = ConfusionMatrixDisplay(confusion_matrix=cm_distilbert, display_labels=["Negative", "Positive"])
disp1.plot(ax=axes[0], cmap="Blues", colorbar=False)
axes[0].set_title("DistilBERT confusion matrix")

# RoBERTa
disp2 = ConfusionMatrixDisplay(confusion_matrix=cm_roberta, display_labels=["Negative", "Positive"])
disp2.plot(ax=axes[1], cmap="Blues", colorbar=False)
axes[1].set_title("RoBERTa confusion matrix")

plt.tight_layout()
plt.show()
```
